```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = [16,4]
```

```python
PATH='data/rossmann'
store_df = pd.read_csv(f'{PATH}/store.csv')
sales_df = pd.read_csv(f'{PATH}/sales.csv', dtype={'StateHoliday':np.object}, parse_dates=['Date'])
```

```python
# Number of stores per store type
store_df.groupby('StoreType')['Store'].count()
```

```python
# Number of stores per assortment
store_df.groupby('Assortment')['Store'].count()
```

```python
# Number of stores with competitors within 5 km
store_df.loc[store_df.CompetitionDistance < 5000]['Store'].count()
```

```python
# Distance to the nearest competior: min, avg, max
comp_dist = store_df['CompetitionDistance']
# Format output
f'min: {comp_dist.min()}, avg: {comp_dist.mean()}, max: {comp_dist.max()}'
```

```python
# Historgram with the distance to the nearest competitor
comp_dist.hist();
```

```python
# Sales per store: min, avg, max, median
sales_per_store = sales_df.groupby('Store')['Sales'].sum()
f'min: {sales_per_store.min()}, avg: {sales_per_store.mean()}, max: {sales_per_store.max()}'
```

```python
# Plot sum of sales and number of customers per day of week
sales_df.groupby('DayOfWeek')[['Sales','Customers']].sum().plot.bar(); #secondary_y='Customers'
```

```python
# Boxplot with sales per weekday
sales_df.boxplot(column=['Sales'], by='DayOfWeek');
```

```python
# Sales per store: histgram
sales_per_store.hist(bins=50);
```
